# **Dataset Information**
Source: Data.gov/https://catalog.data.gov/dataset/air-quality-history

Year / Timeline: 2024/2025

The dataset includes numerical features such as AQI, Arithmetic Mean, and Observation Count,

categorical features such as State Name, City Name, and Parameter Name,

```python
# Fill Missing values in 'Method_Code' with the most common value(mode)
df['METHOD_CODE']= df['METHOD_CODE'].fillna(df['METHOD_CODE'].mode()[0])
```

```python
df.dropna(subset=['POLLUTANT_STANDARD'],inplace=True)
```

```python
df.dropna(subset=['EVENT_TYPE'],inplace=True)
```

```python
df['AQIE']= df['AQI'].fillna(df['AQI'].mode()[0])
```

```python
df.isna().sum()
```

**Handling Outliers**

```python
df.iloc[:, 3:].describe(include='number').astype(int)
```

```python
AQI_threshold =df['AQI'].quantile(0.995)
AQI_threshold
```

```python
METHOD_CODE_threshold =df['METHOD_CODE'].quantile(0.995)
METHOD_CODE_threshold
```

```python
OBJECTID_threshold =df['OBJECTID'].quantile(0.995)
OBJECTID_threshold
```

```python
df[df['OBJECTID']>OBJECTID_threshold]
```

```python
df= df[df['OBJECTID']<= OBJECTID_threshold]
```

```python
for column in ['AQI']:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```
